```python
#Question 3
#What is the largest prime factor of the number 600851475143?

import math
```

```python
def is_prime(num):
    count = 3
    prime = num%count != 0

    while prime and count < math.sqrt(num):
        count += 2
        prime = num%count != 0
    return prime

max_val = 600851475143
sqrt_max = math.sqrt(max_val)
sqrt_max = math.floor(sqrt_max) - 3
prime_factors = []

for fact in range(3,sqrt_max,2):
    max_fact = max_val/fact
    if math.floor(max_fact) == max_fact: #fact is actually a factor
        print(f"Fact: {fact}, max_fact: {max_fact}")
        last_digit = max_fact%10
        if max_fact % 3 != 0 and max_fact % 7 != 0 and max_fact % 9 != 0:
            if is_prime(fact):    
                prime_factors.append(fact)
            if is_prime(max_fact):
                prime_factors.append(max_fact)

print(prime_factors)
```

```python
#Question 4
#Find the largest palindrome made from the product of two-digit numbers.

def is_palindrome(num):
    power = 0
    ret = False
    
    while num > 10 ** (power + 1):
        power += 1
        
    if power > 0: #at least 2 digits
        left = power #position of highest digit
        right = 0
        is_palindrome = True

        #check digits until left position >= right position or the palindrome pattern is broken
        while right < left and is_palindrome:
            is_palindrome = (num//10**left) % 10 == (num//10**right) % 10
            left -= 1
            right += 1

        ret = is_palindrome
            
    return ret

############################
a = 999
b = 999

while not is_palindrome(a*b):
    b -= 1
    if b == 899:
        a -= 1
        b = 999

print(a,b,a*b)
```
